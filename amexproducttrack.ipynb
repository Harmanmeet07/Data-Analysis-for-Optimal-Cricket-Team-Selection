CALCULATING TOTAL POINTS FOR EACH BATSMAN BASED ON STRIKE RATE,AVERAGE OF RUNS,100S AND 50S

```python
# Convert necessary columns to numeric or datetime format
batsman['runs'] = pd.to_numeric(batsman['runs'], errors='coerce')
batsman['runs'] = batsman['runs'].fillna(0)
batsman['strike_rate'] = pd.to_numeric(batsman['strike_rate'], errors='coerce')
batsman['strike_rate'] = batsman['strike_rate'].fillna(0)
batsman['match_dt'] = pd.to_datetime(batsman['match_dt'], errors='coerce')
```

```python
# Calculate the total runs for each batsman
total_runs = batsman.groupby(['batsman_id', 'batsman'])['runs'].sum().reset_index()
total_runs.columns = ['batsman_id', 'batsman', 'total_runs']
```

```python
#strike rate
batsman_strike_rate = batsman.groupby(['batsman_id','batsman'])['strike_rate'].mean().reset_index()
```

```python
batsman_strike_rate
```

```python
# Calculating the average runs for each batsman
average_runs = batsman.groupby(['batsman_id', 'batsman'])['runs'].mean().reset_index()
average_runs.columns = ['batsman_id', 'batsman', 'average_runs']
```

```python
# Calculating the number of centuries and half-centuries
centuries = batsman[batsman['runs'] >= 100].groupby(['batsman_id', 'batsman']).size().reset_index(name='centuries')
half_centuries = batsman[(batsman['runs'] >= 50) & (batsman['runs'] < 100)].groupby(['batsman_id', 'batsman']).size().reset_index(name='half_centuries')
```

```python
# Merging
metrics = pd.merge(total_runs, average_runs, on=['batsman_id', 'batsman'])
metrics = pd.merge(metrics, centuries, on=['batsman_id', 'batsman'], how='left').fillna(0)
metrics = pd.merge(metrics, half_centuries, on=['batsman_id', 'batsman'], how='left').fillna(0)
metrics=pd.merge(metrics,batsman_strike_rate,on=['batsman_id','batsman'],how='left').fillna(0)
```

```python
#defining functions to allot points according to the point system

# Function to calculate points based on the strike rate
def strike_rate_points(sr):
    if sr >= 150:
        return 50
    elif sr >= 100:
        return 40
    elif sr >= 80:
        return 30
    else:
        return 0

# Function to calculate points based on the average runs
def average_points(avg):
    if avg >= 50:
        return 30
    elif avg >= 40:
        return 20
    elif avg >= 30:
        return 10
    else:
        return 5

# Function to calculate points for centuries
def centuries_points(cent):
    if cent >= 3:
        return 30
    elif cent == 2:
        return 20
    elif cent == 1:
        return 10
    else:
        return 0

# Function to calculate points for half-centuries
def half_centuries_points(hc):
    if hc >= 5:
        return 20
    elif hc >= 3:
        return 10
    elif hc >= 1:
        return 5
    else:
        return 0
```

```python
#applying points
batsman_strike_rate['strike_rate_points'] = batsman_strike_rate['strike_rate'].apply(strike_rate_points)
metrics = pd.merge(metrics, batsman_strike_rate[['batsman_id', 'strike_rate_points']], on='batsman_id', how='left')
```

```python
#applying points
metrics['average_points'] = metrics['average_runs'].apply(average_points)
metrics['centuries_points'] = metrics['centuries'].apply(centuries_points)
metrics['half_centuries_points'] = metrics['half_centuries'].apply(half_centuries_points)
```

```python
# Calculating the total points
metrics['total_points'] = (metrics['strike_rate_points'] + metrics['average_points'] +
                           metrics['centuries_points'] + metrics['half_centuries_points'])
```

CALCULATING PERFORMANCE SCORE OF PLAYERS

```python
batsman_cumulative_runs = batsman.groupby('batsman_id').agg(
    cumulative_runs=pd.NamedAgg(column='runs', aggfunc='sum')
).reset_index()
```

```python
batsman_cumulative_runs
```

```python
#normalizing runs for meaningful comparisons
min_run =batsman_cumulative_runs['cumulative_runs'].min()
max_run = batsman_cumulative_runs['cumulative_runs'].max()
```

```python
batsman_cumulative_runs['normalized_runs'] = 1 + ((batsman_cumulative_runs['cumulative_runs'] - min_run) * (100 - 1) / (max_run - min_run))
```

```python
batsman_cumulative_runs
```

```python
batsman_data = batsman_cumulative_runs.merge(
    batsman[['batsman_id','runs', 'match_dt','bowler_id']],
    on='batsman_id',
    how='left'
)
```

Consistency:

Low Standard Deviation: This suggests that the batsman is consistent in their performance, regularly scoring around the same number of runs.

High Standard Deviation:  This suggests inconsistency, where the batsman may have very high scores in some matches and very low scores in others.

```python
batsman_std = batsman_data.groupby('batsman_id').agg(
    std=pd.NamedAgg(column='runs', aggfunc=np.std)
).reset_index()
```

```python
batsman_std
```

```python
batsman_std['inverse_std'] = batsman_std['std'].apply(lambda x: 1 / x if x != 0 else np.inf)
```

```python
batsman_std.rename(columns={'inverse_std': 'consistency'}, inplace=True)
```

```python
batsman_std
```

```python
min_consistency = batsman_std['consistency'].min()
max_consistency = batsman_std['consistency'].max()
```

```python
batsman_std['normalized_consistency'] = 1 + ((batsman_std['consistency'] - min_consistency) * (100 - 1) / (max_consistency - min_consistency))

print(batsman_std)
```

RECENCY

Here we are using 'calculate_recency' function to calculates a weighted average of a batsman's runs, giving more weight to recent matches. This helps assess the batsman's current form by emphasizing recent performances over older ones.

```python
def calculate_recency(group):
    group = group.sort_values('match_dt')
    decay_rate = 0.1

    group['match_dt'] = pd.to_datetime(group['match_dt'])
    time_diff = (group['match_dt'].max() - group['match_dt']).dt.days
    weights = np.exp(-decay_rate * time_diff)
    return np.average(group['runs'], weights=weights)

batsman_recency = batsman_data.groupby('batsman_id').apply(calculate_recency).reset_index(name='recency')
```

```python
min_recency = batsman_recency['recency'].min()
max_recency = batsman_recency['recency'].max()
```

```python
batsman_recency['normalized_recency'] = 1 + ((batsman_recency['recency'] - min_recency) * (100 - 1) / (max_recency - min_recency))

print(batsman_recency)
```

```python
batsman_performance = batsman_cumulative_runs.merge(
    batsman_std, on='batsman_id'
).merge(
    batsman_recency, on='batsman_id'
)
```

```python
batsman_performance
```

```python
batsman_performance['performance_score'] = (
    0.5 * batsman_performance['normalized_runs'] +
    0.3 * (batsman_performance['normalized_consistency'].fillna(0)) +
    0.2 * batsman_performance['normalized_recency']
)
```

```python
#calculating total balls bowled and total wickets taken
total_ballsbowled = bowler.groupby('bowler_id')['balls_bowled'].sum()
total_wickets = bowler.groupby('bowler_id')['wicket_count'].sum()
```

```python
# Computing strike rate for each bowler
bowler_strike_rate = (total_ballsbowled / total_wickets).reset_index()
bowler_strike_rate.columns = ['bowler_id', 'strike_rate']
bowler_strike_rate['strike_rate'].replace([np.inf, -np.inf], np.nan, inplace=True)
print(bowler_strike_rate)
```

```python
#average economy
bowler_economy= bowler.groupby(['bowler_id', 'bowler'])['economy'].mean().reset_index()
```

```python
bowler_economy
```

  We are not clearly given the runs conceded so we can calculate it by the formula of economy.



```python
bowler['runs_conceded'] = bowler['economy'] * bowler['balls_bowled'] / 6
print(bowler[['bowler_id', 'bowler', 'balls_bowled', 'economy', 'runs_conceded']])
```

```python
total_runs_conceded = bowler.groupby('bowler_id')['runs_conceded'].sum()
total_wickets = bowler.groupby('bowler_id')['wicket_count'].sum()
```

```python
#calculating average
bowler_average = (total_runs_conceded / total_wickets).reset_index()
bowler_average.columns = ['bowler_id', 'bowler_average']
bowler_average['bowler_average'].replace([np.inf, -np.inf], np.nan, inplace=True)
print(bowler_average)
```

```python
#Calculating 4W_Per_Innings
wickets_4w = bowler[bowler['wicket_count'] >= 4]

# Group by bowler and count the number of innings with 4 or more wickets
four_w_per_innings = wickets_4w.groupby('bowler_id').size().reset_index(name='4W_Per_Innings')
```

```python
four_w_per_innings
```

```python
bowler_score_df = pd.merge(bowler_strike_rate, bowler_economy, on=['bowler_id'], how='outer')
bowler_score_df = pd.merge(bowler_score_df, bowler_average, on=['bowler_id'], how='outer')
bowler_score_df = pd.merge(bowler_score_df, four_w_per_innings, on=['bowler_id'], how='outer')
```

```python
bowler_score_df
```

```python
#Giving them points based on the given criteria
# Function to calculate points based on the strike rate
def strike_rate_points(sr):
    if sr <= 15:
        return 30
    elif sr <= 19:
        return 20
    elif sr <= 24:
        return 10
    else:
        return 0

# Function to calculate points based on the economy rate
def economy_points(econ):
    if econ <= 3:
        return 50
    elif econ <= 5:
        return 40
    elif econ <= 7:
        return 30
    else:
        return 0

# Function to calculate points based on the bowling average
def average_points(avg):
    if avg <= 20:
        return 30
    elif avg <= 30:
        return 20
    elif avg <= 40:
        return 10
    else:
        return 0

# Function to calculate points for 4-wicket hauls
def four_wicket_hauls_points(fwh):
    if fwh >= 4:
        return 30
    elif fwh >= 2:
        return 20
    elif fwh >= 1:
        return 10
    else:
        return 0
```

```python
#applying points
bowler_score_df['strike_rate_points'] = bowler_score_df['strike_rate'].apply(strike_rate_points)
bowler_score_df['economy_points'] = bowler_score_df['economy'].apply(economy_points)
bowler_score_df['average_points'] = bowler_score_df['bowler_average'].apply(average_points)
bowler_score_df['four_wicket_hauls_points'] = bowler_score_df['4W_Per_Innings'].apply(four_wicket_hauls_points)
```

```python
# Calculate total points for each bowler
bowler_score_df['total_points'] = (bowler_score_df['strike_rate_points'] +
                             bowler_score_df['economy_points'] +
                             bowler_score_df['average_points'] +
                             bowler_score_df['four_wicket_hauls_points'])

# Printing the final DataFrame with points calculated
print(bowler_score_df[['bowler_id', 'bowler', 'strike_rate', 'economy', 'bowler_average', '4W_Per_Innings', 'total_points']])
```
